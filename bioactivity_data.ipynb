```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

Search for Target Protein

```python
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

Select and retrieve bioactivity data for Replicase polyprotein 1a

```python
selected_target = targets.target_chembl_id[8]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

IC50 = Concentration to elicit 50% inhibition of target protein

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

```python
# Save to csv
df.to_csv('bioactivity_data.csv', index=False)
```

Handle missing data

```python
df2 = df[df.standard_value.notna()]
df2
```

Data pre-processing

```python
# Label compounds as either active, inactive, or intermediate
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
```

```python
df3 = df2[["molecule_chembl_id", "canonical_smiles", "standard_value"]]
df4 = df3.join(pd.Series(bioactivity_class, name="bioactivity_class"), how='inner')
df4
```

```python
df4.to_csv('data/bioactivity_preprocessed_data.csv', index=False)
```
